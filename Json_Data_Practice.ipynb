```python
import json
import pandas as pd
from pandas import json_normalize

# Open a JSON file and load its content into a Python dictionary
with open("sample.json", mode="r") as file:
  data = json.load(file)

print(data)
```

```python
# Read JSON file directly into a Pandas DataFrame
df = pd.read_json("sample.json")

print(df.head())
```

```python
#Retrieves Order Id and total after making nested json Flatten
df_orders = json_normalize(data, record_path=['orders'])

print(df_orders[['order_id', 'total']])
```

```python
#Retrieves customer_id, order_id, product, quantity, total
df_data = pd.json_normalize(data, record_path=['orders', 'items'], meta=[['orders', 'total'],['orders','order_id'], 'customer_id'])

print(df_data[['customer_id', 'orders.order_id', 'product', 'quantity', 'orders.total']])
```

```python
df_data.to_json("processed_orders.json", orient="records", indent=4)
```
