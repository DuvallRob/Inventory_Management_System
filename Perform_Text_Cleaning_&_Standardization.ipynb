```python
import re
import unicodedata
import ftfy
import contractions
import string
import nltk
from nltk.corpus import stopwords

f = open('text_sample.txt', 'r')
content = f.read()
fixed_text = ftfy.fix_text(content)
normalized_text = unicodedata.normalize('NFKD', fixed_text).encode('ascii', 'ignore').decode()
expanded_text = contractions.fix(normalized_text)
clean_text = expanded_text.replace(',', '').replace('*', '').replace('  ', '').replace('div', '').replace('/','').replace('<>','').replace('spacesspecial', 'spaces special').replace('!', '.').replace('inconsistent', '').replace("\n\n", '\n')
clean_text = re.sub("  ", " ", clean_text)

clean_text = clean_text.translate(str.maketrans('', '', string.punctuation))

# Standardizing common terms
common_terms = {
    "Dr.": "Doctor",
    "NYC": "New York",
    "N.Y.": "New York",
    "USD": "dollars",
    "$": "dollars"
}

def standardize_terms(text, replacements):
    for key, value in replacements.items():
        text = text.replace(key, value)
    return text

clean_text = standardize_terms(clean_text, common_terms)

print(clean_text)
```
